```python
def intra_city_bus_journey(df, value):
    """Function to drop last two columns of dataframe and obtain average prices for intra-city bus journey of a state.
        
        Args:
            df (Dataframe): Dataframe belonging to a state
            
        
        Returns:
            Dataframe: Dataframe containing average prices for all intra-city bus journey ranging from 2016 to 2020
    """
    df = df.copy()
    # select first row
    df1 = df.iloc[[0]]
    # exclude last two columns
    df1 = df1.iloc[:,:-2]
    # Insert new column with an input value
    df1.insert(0,"State", value)
    return df1
```

```python
def airfare(df, value):
    """Function to drop last two columns of dataframe and obtain average airfare prices for specified routes in states with domestic airports.
        
        Args:
            df (Dataframe): Dataframe belonging to a state
            
        
        Returns:
            Dataframe: Dataframe containing average airfare prices for specified routes from 2016 to 2020
    """
    df = df.copy()
    # select third row
    df1 = df.iloc[[2]]
    # exclude last two columns
    df1 = df1.iloc[:,:-2]
    # Insert new column with an input value
    df1.insert(0,"State", value)
    return df1
```

```python
def water_journey(df, value):
    """Function to drop last two columns of dataframe and obtain average prices for water journey within a state.
        
        Args:
            df (Dataframe): Dataframe belonging to a state
            
        
        Returns:
            Dataframe: Dataframe containing average prices for all water journey ranging from 2016 to 2020
    """
    df = df.copy()
    # select fifth row
    df1 = df.iloc[[4]]
    # exclude last two columns
    df1 = df1.iloc[:,:-2]
    # Insert new column with an input value
    df1.insert(0,"State", value)
    return df1
```

```python
def calculate_percentage(df):
    """Function to calculate the percentage of the transport fare for all states per year
        
        Args:
            df (Dataframe): Dataframe of all states
            
        
        Returns:
            Dataframe: A dataframe with yearly returns of the percentage transport fare for all states from 2016 to 2020
    """
    # select a range of columns 
    df_2016 = df.loc[: , "Jan-16":"Dec-16"]
    df_2017 = df.loc[: , "Jan-17":"Dec-17"]
    df_2018 = df.loc[: , "Jan-18":"Dec-18"]
    df_2019 = df.loc[: , "Jan-19":"Dec-19"]
    df_2020 = df.loc[: , "Jan-20":"Mar-20"]
    
    # use this range of columns selected to obtain the mean
    df['2016'] = df_2016.mean(axis=1)
    df['2017'] = df_2017.mean(axis=1)
    df['2018'] = df_2018.mean(axis=1)
    df['2019'] = df_2019.mean(axis=1)
    df['2020'] = df_2020.mean(axis=1)
    
    # calculate the percentage change 
    df['percentage-change-16-18'] = abs(df['2016'] - df['2018'])/df['2016'] * 100
#     df['percentage-change-17-18'] = abs(df['2017'] - df['2018'])/df['2017'] * 100
#     df['percentage-change-18-19'] = abs(df['2018'] - df['2019'])/df['2018'] * 100
    df['percentage-change-18-20'] = abs(df['2018'] - df['2020'])/df['2018'] * 100

    return df
```

```python
# calculate the percentage for each route in all states

states_intra_city_bus_journey_df = calculate_percentage(states_intra_city_bus_journey_df)
states_inter_city_bus_journey_df = calculate_percentage(states_inter_city_bus_journey_df)
states_airfare_df = calculate_percentage(states_airfare_df)
states_motorcycle_journey_df = calculate_percentage(states_motorcycle_journey_df)
states_water_journey_df = calculate_percentage(states_water_journey_df)
```

```python
def read_as_csv(df, file_name):
    """Function to read dataframe into CSV file
        
        Args:
            df (Dataframe): Dataframe of all states
            
        
        Returns:
            file: CSV file
    """
    df = df.to_csv(file_name, index=False)

    return df
```
